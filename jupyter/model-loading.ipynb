```python
architectures = ["densenet121", "efficientnet_b1", "efficientnet_v2_s", "resnet50"]
```

```python
classes = ["chexpert", "chexternal", "chexternal_pneumo", "pneumonia"]
n_classes = [13, 5, 6, 1]
```

```python
base = "//hastur/scratch/hekalo/Models/labels_chexpert/bce/"
```

```python
for i, labels in enumerate(classes):
    for architecture in architectures:
       model = sequential_model(architecture, n_classes=n_classes[i])
       path = base + "/" + labels + "/" + architecture + "/" + architecture + "_best"
       state = torch.load(path)
       model.load_state_dict(state["model"])
       torch.save(model, path + ".pt")
```
